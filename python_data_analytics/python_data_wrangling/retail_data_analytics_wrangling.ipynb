## Finding RFM

RFM is a method used for analyzing customer value. It is commonly used in database marketing and direct marketing and has received particular attention in the retail and professional services industries.

RFM stands for three dimensions:

- Recency – How recently did the customer purchase?

- Frequency – How often do they purchase?

- Monetary Value – How much do they spend?



```python
recency = retail_df.groupby(by='CustomerID')['Quantity'].sum()
recency
```

```python
invoice = retail_df.groupby(by='CustomerID')['Invoice'].count().reset_index()
invoice_count = invoice['Invoice']
invoice_count
```

```python
monetary = retail_df.groupby(by='CustomerID')['Sales'].sum().reset_index()
monetary_sum = monetary['Sales']
monetary_sum
```

```python
rfm_df = recency.reset_index()
rfm_df = rfm_df.rename(columns={'Quantity': 'Recency'})
rfm_df['Frequency'] = invoice_count
rfm_df['Monetary'] = monetary_sum
rfm_df
```

# RFM Segmentation

```python
rfm_df['RecencyScore'] = pd.qcut(rfm_df['Recency'],5,labels=[5,4,3,2,1])
rfm_df['FrequencyScore'] = pd.qcut(rfm_df['Frequency'].rank(method="first"),5,labels=[1,2,3,4,5])
rfm_df['MonetaryScore'] = pd.qcut(rfm_df['Monetary'],5,labels=[1,2,3,4,5])
rfm_df
```

```python
rfm_df["RFM_SCORE"] = rfm_df['RecencyScore'].astype(str) + rfm_df['FrequencyScore'].astype(str) + rfm_df['MonetaryScore'].astype(str)
rfm_df
```

```python
#segmenting of customers according to RecencyScore and FrequencyScore values
seg_map = {
    r'[1-2][1-2]': 'Hibernating',
    r'[1-2][3-4]': 'At Risk',
    r'[1-2]5': 'Can\'t Lose',
    r'3[1-2]': 'About to Sleep',
    r'33': 'Need Attention',
    r'[3-4][4-5]': 'Loyal Customers',
    r'41': 'Promising',
    r'51': 'New Customers',
    r'[4-5][2-3]': 'Potential Loyalists',
    r'5[4-5]': 'Champions'
}
```

```python
#creation of segment variable
rfm_df['Segment'] = rfm_df['RecencyScore'].astype(str) + rfm_df['FrequencyScore'].astype(str)
rfm_df['Segment'] = rfm_df['Segment'].replace(seg_map, regex=True)
```
